#### Problem : We notice that the comments are written in different ways: Arabic, French, English and Tunisian dialect also we notice the use of latin alphabet and arabic alphabet either separately or combined

#### Solution : We need to detect the language of the comment, translate the text from Arabic and French to English and create a dictionnary for the Tunisian Dialect

```python
#language detection 
import spacy
from langdetect import detect
import langdetect
```

```python
Comment_col
```

```python
languages_langdetect = []

# the try except blook because there is some tweets contain links
for comment in Comment_col:
    try:
        result = detect(comment)
        if (result != 'ar' and result != 'fr' and result != 'en'): #other then ar fr and english are in tn 
            result='tn'
        
    except:
        result = 'unknown'
    
    finally:
        languages_langdetect.append(result)
```

#### the comments with languages that are defined as other then arabic, french and english are in tunisian dialect therefore we defined them as tn : tunisian dialect in the languages column

```python
languages_langdetect
```

```python
comments['language'] = languages_langdetect
comments
```

#### Problem: There are missing values 
#### Solution: We need to replace missing values

```python
missing_values= ["NA","",None, np.NAN]
missing= comments.isin(missing_values)
comments_filled=comments.mask(missing,"missing")
comments_filled
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(comments_filled.index,comments_filled.nbreacts)
plt.title('Number of reacts for each comment')
plt.show()
```

```python
plt.plot(comments_filled.postid,comments_filled.nbreacts)
plt.title('number of reacts of each post')
plt.show()
```

```python
arabic = comments.loc[comments['language']=='ar'].count()[0]
french = comments.loc[comments['language']=='fr'].count()[0]
english = comments.loc[comments['language']=='en'].count()[0]
tunisian = comments.loc[comments['language']=='tn'].count()[0]
unknown = comments.loc[comments['language']=='unknown'].count()[0]
labels = ['Arabic', 'French', 'English', 'Tunisian Dialect', 'Unknown']
plt.figure(figsize=(10, 10))
plt.pie([arabic, french, english, tunisian, unknown],labels = labels, shadow = True, autopct='%1.1f%%')
plt.title('Distribution of languages in comments')
plt.legend()
plt.show()
```
